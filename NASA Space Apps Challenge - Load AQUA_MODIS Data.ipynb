```python
def replace_missing_data(array):
    array_new = array.copy()
    for row in range(len(array)):
        for col in range(len(array[row])):
            if array[row, col] == 99999.0:
                array_new[row, col] = np.nan
    return array_new
```

```python
directory = 'AQUA_MODIS_monthly/'

for filename in tqdm(os.listdir(directory)):
    #print(filename)
    path_ = os.path.join(directory, filename)
    try:
        with open(path_, 'r') as f:
            file = list(csv.reader(f, delimiter=","))
        data = np.array(file, dtype=np.float)
        data_cropped = data[y0:y1,x0:x1]
        data_cropped = replace_missing_data(data_cropped)
        np.savetxt(f'Processed/{filename}', data_cropped, delimiter=",")
    except: 
        print(filename)
```
